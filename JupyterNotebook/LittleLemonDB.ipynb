```python
#Create check booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()

    drop_procedure = """
        DROP PROCEDURE IF EXISTS `CheckBooking`;
    """
    cursor.execute(drop_procedure)
    
    #Execute query
    create_checkbooking = """            
            CREATE PROCEDURE `CheckBooking`(IN booking_date DATE, IN table_no INT)
            BEGIN
                DECLARE checkRes BOOLEAN;
                
                SELECT COUNT(*) INTO checkRes FROM Bookings
                WHERE BookingDate = booking_date AND TableNumber = table_no;

                IF checkRes THEN
                    SELECT CONCAT('The date ', booking_date, ' and table ', table_no, ' are already booked.') AS Message;
                ELSE 
                    SELECT CONCAT('Table ', table_no, ' is available on ', booking_date, '.') AS Message;
                END IF;    
            END
    """
    cursor.execute(create_checkbooking)

    print("MySQL procedure successfully created")

    cursor.close()

except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```

```python
#Create update booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()

    drop_procedure = """
        DROP PROCEDURE IF EXISTS `UpdateBooking`;
    """
    cursor.execute(drop_procedure)
    
    #Execute query
    create_updatebooking = """            
            CREATE PROCEDURE `UpdateBooking`(IN booking_id INT, IN date DATE)
            BEGIN
                UPDATE Bookings
                SET BookingDate = date
                WHERE BookingID = booking_id;
            END
    """
    cursor.execute(create_updatebooking)

    print("MySQL procedure successfully created")

    cursor.close()

except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```

```python
#Using update booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()

    #Asking for user-input
    booking_id = input("Please enter your Booking ID: ")
    new_booking_date = input("Please enter the new Booking Date (YYYY-MM-DD): ")

    # Convert Booking ID to integer if necessary
    booking_id = int(booking_id)
    
    #SQL query to check previous booking date & execute
    booking = """
    SELECT BookingID, BookingDate, TableNumber FROM Bookings WHERE BookingID = %s;
    """
    cursor.execute(booking, (booking_id,))
    booking_results1 = cursor.fetchall()
    for i in booking_results1:
        print(f"\nYou are requesting a change for Booking ID #{i[0]}.\nYour previous booking date was: {i[1]}")

    #Calling procedure to update
    update_booking = """            
            CALL UpdateBooking(%s, %s)
    """
    print("\nInitating booking update...\n")
    
    cursor.execute(update_booking, (booking_id, new_booking_date))
    
    confirm = input("Do you want to complete the booking? (y/n) ")
    if confirm == 'y':
        connection.commit()
        
        cursor.execute(booking, (booking_id,))
        booking_results2 = cursor.fetchall()
        for i in booking_results2:
            print(f"Your new booking date is {i[1]}")
            
    else:
        print("Your booking update has been cancelled.")    

    cursor.close()
    
except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```

```python
#Create add booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()
    
    #Drops existing procedure under the same name
    drop_procedure = """
        DROP PROCEDURE IF EXISTS `AddBooking`;
    """
    cursor.execute(drop_procedure)
    
    #Execute query
    create_addbooking = """            
            CREATE PROCEDURE `AddBooking`(IN date DATE, IN table_no INT)
            BEGIN
                DECLARE check_res BOOLEAN;
                
                START TRANSACTION;
                    
                SELECT COUNT(*) INTO check_res FROM Bookings WHERE BookingDate = date AND TableNumber = table_no;

                IF check_res THEN
                    ROLLBACK;
                    SELECT "The date and table is already booked.";
                        
                ELSE
                    INSERT INTO Bookings (BookingDate, TableNumber) VALUES (date, table_no); 
                    COMMIT;

                END IF;
            END
    """
    cursor.execute(create_addbooking)

    print("MySQL procedure successfully created")

    cursor.close()

except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```

```python
#Using add booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()

    #Asking for user-input
    table_no = input("Please enter your desired table number: ")
    booking_date = input("Please enter your desired Booking Date (YYYY-MM-DD): ")

    #Convert Booking ID to integer if necessary
    table_no = int(table_no)
    
    #SQL query to check previous booking date & execute
    booking = """
    SELECT BookingID, BookingDate, TableNumber FROM Bookings WHERE BookingDate = %s;
    """
    cursor.execute(booking, (booking_date,))
    booking_results1 = cursor.fetchall()
    
    print("\nChecking Booking...\n")
    
    print(f"On {booking_date}, the tables booked are: ")
    for i in booking_results1:
        print(f"Table number: {i[2]}")

    #Calling procedure to update
    add_booking = """            
            CALL AddBooking(%s, %s)
    """
    cursor.execute(add_booking, (booking_date, table_no))
    check_booking = cursor.fetchall()
    for i in check_booking:
        print(i[0])

    if i[0] == 'The date and table is already booked.':
        print("The add booking process has been cancelled.")
    
    else:
        confirm = input("\nLooks like the table is available on that date! Do you want to complete the booking? (y/n)")
        if confirm == 'y':
            connection.commit()
            
            cursor.execute(booking, (booking_date,))
            booking_results2 = cursor.fetchall()
            for i in booking_results2:
                new_table = i[-1]
            print(f"\nAwesome, your new booking information is:\nDate: {booking_date}\nTable Number: {new_table}")
                
        else:
            print("The add booking process is cancelled.")    
    
    cursor.close()
    
except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```

```python
#Create cancel booking procedure

try:
    #Establish connection to server
    connection = connector.connect(
        user = user, 
        password = password,
        db = db)

    #Create cursor object
    cursor = connection.cursor()

    #Drops existing procedure under the same name
    drop_procedure = """
        DROP PROCEDURE IF EXISTS `CancelBooking`;
    """
    cursor.execute(drop_procedure)
    
    #Execute query
    create_cancelbooking = """            
            CREATE PROCEDURE `CancelBooking`(IN booking_id INT)
            BEGIN
               DELETE FROM Bookings
               WHERE BookingID = booking_id;
            END
    """
    cursor.execute(create_cancelbooking)

    print("MySQL procedure successfully created")

    cursor.close()

except connector.Error as err:
    print(f"Error: {err}")

finally:
    if connection.is_connected():
        connection.close()
```
